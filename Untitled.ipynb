```python
import cv2
import numpy as np

def calculate_metallic(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Check if the image was loaded properly
    if image is None:
        print(f"Error: Could not load image at {image_path}")
        return None
    
    # Convert the image to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # Define the range for metallic colors in HSV
    lower_metallic = np.array([0, 0, 200])  # High brightness, low saturation
    upper_metallic = np.array([180, 55, 255])
    
    # Create a mask for metallic colors
    metallic_mask = cv2.inRange(hsv_image, lower_metallic, upper_metallic)
    
    # Calculate the percentage of metallic pixels
    metallic_percentage = np.sum(metallic_mask > 0) / metallic_mask.size * 100
    
    # Display the mask
    cv2.imshow('Metallic Mask', metallic_mask)
    cv2.waitKey(0)  # Wait for a key press to close the window
    cv2.destroyAllWindows()
    
    return metallic_percentage

def compare_metallic(image_path1, image_path2):
    metallic1 = calculate_metallic(image_path1)
    metallic2 = calculate_metallic(image_path2)
    
    if metallic1 is None or metallic2 is None:
        return
    
    if metallic1 > metallic2:
        print(f"The first image ({image_path1}) has more metallic color with {metallic1:.2f}% metallic pixels.")
    elif metallic2 > metallic1:
        print(f"The second image ({image_path2}) has more metallic color with {metallic2:.2f}% metallic pixels.")
    else:
        print(f"Both images have the same percentage of metallic pixels: {metallic1:.2f}%.")

# Example usage
# Paths to the images
image_path1 = 'normal.jpeg'
image_path2 = 'defect2.jpg'
compare_metallic(image_path1, image_path2)
```

```python
import numpy as np
from PIL import Image

def resize_image(image, target_size):
    return image.resize(target_size, Image.ANTIALIAS)

def image_difference(image_path1, image_path2):
    # Load images
    img1 = Image.open(image_path1)
    img2 = Image.open(image_path2)

    # Resize images to have the same dimensions
    target_size = (min(img1.width, img2.width), min(img1.height, img2.height))
    img1_resized = resize_image(img1, target_size)
    img2_resized = resize_image(img2, target_size)

    # Convert images to numpy arrays
    arr1 = np.array(img1_resized)
    arr2 = np.array(img2_resized)

    # Calculate absolute difference between the two images
    diff = np.abs(arr1 - arr2)

    # Calculate percentage difference based on maximum possible pixel difference
    max_diff = 255  # for 8-bit images (0-255)
    difference_percentage = (np.sum(diff) / (arr1.size * max_diff)) * 100

    # Similarity percentage
    similarity_percentage = 100 - difference_percentage

    return similarity_percentage, difference_percentage

# Example usage:
image_path1 = 'normal3.jpg'
image_path2 = 'defect3.jpg'
similarity, difference = image_difference(image_path1, image_path2)
print(f"Similarity: {similarity:.2f}%")
print(f"Difference: {difference:.2f}%")
```
